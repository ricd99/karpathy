```python
b = {}

for w in words:
    chs = ["<S>"] + list(w) + ["<E>"]
    for ch1, ch2 in zip(chs, chs[1:]):
        bigram = (ch1, ch2)
        b[bigram] = b.get(bigram, 0) + 1
```

key tells sorted function what value to use when comparing elements for sorting. 
key is called once for each element in the iterable
that element is passed at the argument for key

in the below case, kv recieves output of b.items(). this is each element in the dict.

```python
sorted(b.items(), key = lambda kv: -kv[1]) # sort by dict values, not keys. sort in desc. order
```

```python
import torch
```

```python
N = torch.zeros((28, 28), dtype=torch.int32)
```

```python
chars = sorted(list(set("".join(words))))
stoi = {s:i for i, s in enumerate(chars)}
stoi["<S>"] = 26
stoi["<E>"] = 27
stoi
```

```python
for w in words:
    chs = ["<S>"] + list(w) + ["<E>"]
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        N[ix1, ix2] += 1
```

```python
itos = {i:s for s,i in stoi. items()}
itos
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16, 16))
plt.imshow(N, cmap="Blues")
for i in range(28):
    for j in range(28):
        chstr = itos[i] + itos[j]
        plt.text(j, i, chstr, ha="center", va="bottom", color="gray")          # flipped indexing because matrix is (row, column) while table 9s (horizontal pos, vertical pos)
        plt.text(j, i, N[i, j].item(), ha="center", va="top", color="gray")       
plt.axis("off")
```
